# Exercise: Visualizing Stock Indices

We're loading three stock indices here; the Nasdaq, the SP 500, the Dow Jones. We're also setting it up correclty with date time. 

```python
indices = pd.read_csv('indices.csv')
indices.info()
indices["Date"] = indices["Date"].astype("datetime64")
indices.set_index('Date', inplace=True)
indices 
```

## Task 1: Plot the stock of all three indices in one plpt

```python
indices.plot()
```

## Task 2: Plot The inidices in three sub-plots below each other

Is that easy to compare? 

```python
fig, subfigs = plt.subplots(3, 1, figsize=(16, 10))

# show the markers on the line
subfigs[0].plot(indices["Dow Jones"], label="Dow Jones")
subfigs[1].plot(indices["Nasdaq"], label="Nasdaq")
subfigs[2].plot(indices["SP 500"], label="SP 500")
# add a legend
```

## Task 3: Plotting Change
 
Create a derived dataset that contains the change relative to the first day in your dataset and plot this in one plot. 

```python
indices["Dow Change"] = (indices['Dow Jones'] - indices['Dow Jones'].iloc[0]) / indices['Dow Jones'].iloc[0]
indices["Nasdaq Change"] = (indices['Nasdaq'] - indices['Nasdaq'].iloc[0]) / indices['Nasdaq'].iloc[0]
indices["SP 500 Change"] = (indices['SP 500'] - indices['SP 500'].iloc[0]) / indices['SP 500'].iloc[0]
```

```python
indices[['Dow Change', 'Nasdaq Change', 'SP 500 Change']].plot()
```
